## 3. Pré-processamento

Minimizar ou eliminar problemas existentes em um conjunto de dados, bem como torná-los mais adequados para sua utilização por um determinado algoritmo de aprendizagem de máquina.

## 3.1 Eliminação de atributos irrelevantes

O objetivo do projeto é identificar os produtos com o melhor desconto, que é a (`discount_price`). Não é preciso uma análise profunda para observar que os atributos `id_produto` e `name_product` não contribuem para a identificação. Portanto, em uma tarefa de aprendizado de máquina, devemos remover esses atributos, pois são irrelevantes. 

*Em cenários reais, muitas vezes é necessário consultar especialistas para ajudar a identificar quais atributos são irrelevantes.*

```python
df_dataset_new = df_dataset.drop(columns=['category','rating','rating_count', 'about_product', 'user_id',
                                          'user_name', 'review_id', 'review_title', 'review_content', 'img_link', 'product_link'])

# imprimir o dataframe
display(df_dataset_new.head(n=25))
```

```python
# índices das linhas que contém valores NaN
idxRowNan = pd.isnull(df_dataset_new).any(axis=1).to_numpy().nonzero()

# imprime apenas as linhas com valores ausentes
display(df_dataset_new.iloc[idxRowNan])
```

```python
df_duplicates = df_dataset_new[df_dataset.duplicated(subset=['product_id','product_name','discounted_price','actual_price', 'discount_percentage'],keep=False)] 

# se houver valores redundantes ou inconsistentes, imprima 
if len(df_duplicates) > 0:
    print('\nObjeto(s) redundante(s) e/ou inconsistente(s):')
    display(df_duplicates)
else:
    print('Não existem objetos redundantes e/ou inconsistentes!')
```

```python
def delDuplicatas( df_dataset_new ):
    
    # remove o(s) objeto(s) duplicado(s), mantendo apenas a primeira ocorrência
    df_dataset_normal = df_dataset_new.drop_duplicates(keep = 'first')

    return df_dataset_normal

df_dataset_normal = delDuplicatas( df_dataset_new )
```

```python
df_duplicates = df_dataset_normal[ df_dataset_normal.duplicated(subset=['product_id','product_name','discounted_price','actual_price', 'discount_percentage'],keep=False)] 

# se existir objetos inconsistentes, exibir
if len(df_duplicates)>0:
    print('\nObjeto(s) inconsistente(s):')
    display(df_duplicates)
else:
    print('Não existem objetos redundantes e/ou inconsistentes!')
    
```

## 4.1 Detecção de outliers (valor atípico)
Outro passo importante na análise e tratamento dos dados é a detecção de outliers (dados gerados por leituras incorretas, erros de digitação etc).

`Gráfico Box Plot`

Uma das maneiras mais simples de verificar se os dados contém outliers.

---
Como as colunas são do tipo string, primeiro vamos retirar os cifrôes 

```python
# Remover o símbolo '₹' e as vírgulas, e converter para float, com o .loc para evitar o SettingWithCopyWarning
df_dataset_normal.loc[:, 'discounted_price'] = pd.to_numeric(df_dataset_normal['discounted_price'].replace({'₹': '', ',': ''}, regex=True), errors='coerce')
df_dataset_normal.loc[:, 'actual_price'] = pd.to_numeric(df_dataset_normal['actual_price'].replace({'₹': '', ',': ''}, regex=True), errors='coerce')

# Forçar inferência de tipo de objeto
df_dataset_normal = df_dataset_normal.infer_objects()

# Verificar as mudanças
print(df_dataset_normal[['discounted_price', 'actual_price']].head())
```

```python
print(df_dataset_normal[['discounted_price', 'actual_price', 'discount_percentage']].describe())
```

---
Ao 'transformar' as colunas em números conseguimos visualizar qual o desconto mais aplicado

```python
import matplotlib.pyplot as plt
df_dataset_normal[['discounted_price', 'actual_price']].hist(bins=20, figsize=(10, 6))
plt.show()

df_dataset_normal['discount_percentage'].hist(bins=20, figsize=(10, 60))
plt.show()
```

--- 
Por fim, podemos ver uma análise geral dos preços atuais com os seus desscontos

```python
# Calcular a diferença de preço de forma segura, utilizando .loc
df_dataset_normal.loc[:, 'price_difference'] = df_dataset_normal['actual_price'] - df_dataset_normal['discounted_price']

# Gráfico de dispersão entre o preço original e o preço com desconto
plt.figure(figsize=(10, 6))
plt.scatter(df_dataset_normal['actual_price'], df_dataset_normal['discounted_price'], alpha=0.5)
plt.title('Preço Original vs Preço com Desconto')
plt.xlabel('Preço Original')
plt.ylabel('Preço com Desconto')
plt.show()
```
