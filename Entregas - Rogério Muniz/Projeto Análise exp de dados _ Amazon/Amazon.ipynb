## 5. Tratamento de atributos com valores ausentes
##### Verifica√ß√£o de atributos com valores ausentes na database

```python
# √≠ndices das linhas que cont√©m valores NaN
idxRowNan = pd.isnull(df_dataset).any(axis=1).to_numpy().nonzero()

# imprime apenas as linhas com valores ausentes
display(df_dataset.iloc[idxRowNan])
```

<div>
Ser√° preenchido os valores vazios com a m√©dia de contagem de avalia√ß√£o tendo de acordo a avalia√ß√£o (rating)

```python
def trataFaltantes(df_dataset):

    # Garantir que 'rating_count' seja num√©rico
    df_dataset['rating_count'] = pd.to_numeric(df_dataset['rating_count'], errors='coerce')

    # Calcular a m√©dia de 'rating_count' para cada categoria de 'rating'
    media_por_rating = df_dataset.groupby('rating')['rating_count'].mean()

    # Substituir os valores nulos pela m√©dia correspondente do 'rating'
    df_dataset['rating_count'] = df_dataset.apply(
        lambda row: media_por_rating[row['rating']] if pd.isnull(row['rating_count']) else row['rating_count'], axis=1
    )

    return df_dataset

# Aplicando a fun√ß√£o
df_dataset = trataFaltantes(df_dataset)

# Verificando se ainda h√° valores nulos
print('\nValores nulos restantes:')
print(df_dataset.isnull().sum())
```

## 7. An√°lise dos Dados
A an√°lise das caracter√≠sticas presentes em um conjunto de dados permite a descoberta de padr√µes e tend√™ncias que podem fornecer informa√ß√µes valiosas para compreender o processo que gerou os dados.
Contudo as colunas num√©ricas est√£o sendo interpretadas como strings devido √† presen√ßa de s√≠mbolos. <div>
Para que o describe() funcione corretamente, precisamos converter esses valores para n√∫meros.

```python
# Fun√ß√£o para remover s√≠mbolos e converter para n√∫mero
def limpar_numerico(valor):
    if isinstance(valor, str):  # Verifica se o valor √© uma string
        valor = valor.replace('‚Çπ', '').replace('%', '').strip()  # Remove s√≠mbolos
        return pd.to_numeric(valor, errors='coerce')  # Converte para n√∫mero
    return valor  # Retorna o valor original se n√£o for string

# Aplicar a convers√£o nas colunas relevantes
colunas_para_converter = ['discounted_price', 'actual_price', 'discount_percentage', 'rating']
for coluna in colunas_para_converter:
    df_dataset[coluna] = df_dataset[coluna].apply(limpar_numerico)

# Agora, podemos visualizar as estat√≠sticas corretamente
df_detalhes = df_dataset.describe()
display(df_detalhes)
```

## 8. Detec√ß√£o de outliers (valor at√≠pico)
Outro passo importante na an√°lise e tratamento dos dados √© a detec√ß√£o de outliers (dados gerados por leituras incorretas, erros de digita√ß√£o etc).

- Gr√°fico Box Plot

```python
# gerar um bloxplot para cada atributo
df_dataset.boxplot(figsize=(15,7))
plt.show()
```

---
## 9. Remo√ß√£o de Outliers

```python
# Fun√ß√£o para remover outliers usando o m√©todo IQR
def remove_outliers(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df

# Lista das colunas num√©ricas para remo√ß√£o de outliers
num_columns = ["discounted_price", "actual_price", "discount_percentage", "rating", "rating_count"]

# Aplicar remo√ß√£o de outliers
df_cleaned = remove_outliers(df_dataset, num_columns)
```

---
## 10. Gera√ß√£o do gr√°fico

```python
plt.figure(figsize=(15, 6))
sns.boxplot(data=df_cleaned[num_columns])
plt.xticks(rotation=45)
plt.title("Distribui√ß√£o de pre√ßos, descontos, avalia√ß√µes e contagens de rating ap√≥s limpeza")
plt.show()
```

Verifica√ß√£o se existe rela√ß√£o entre pre√ßo, desconto e avalia√ß√£o.
