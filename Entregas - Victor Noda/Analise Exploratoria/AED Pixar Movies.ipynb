---

## Este notebook tem como objetivo mostrar os seguintes insights:

#### ✔ Sucesso nas Bilheterias: Filmes com maior arrecadação, ROI (Retorno sobre Investimento), ganhos globais vs. domésticos.  
#### ✔ Reconhecimento da Crítica: Filmes da Pixar mais bem avaliados no Rotten Tomatoes, Metacritic e IMDb.  
#### ✔ Análise de Tendências: Crescimento de orçamento, tendências de bilheteria e notas da crítica ao longo do tempo.  
#### ✔ Análise de Correlação: Relação entre orçamento e métricas de sucesso.  
#### ✔ Melhores e Piores Desempenhos: Identificação dos maiores sucessos e fracassos da Pixar.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('pixar_films new.csv', encoding='ascii')

df['release_date'] = pd.to_datetime(df['release_date']) #converter release_date para o formato de datetime

df['ROI'] = (df['box_office_worldwide'] - df['budget']) / df['budget'] #criar uma nova coluna para ROI (return on investment)

sns.set_style("whitegrid")
```

```python
# Plot 1: Top 10 Filmes da Pixar com Maior Bilheteria Mundial
plt.figure(figsize=(12, 6))
top_grossing = df.sort_values(by='box_office_worldwide', ascending=False).head(10)
ax = sns.barplot(y=top_grossing['film'], x=top_grossing['box_office_worldwide'])
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', padding=3)
plt.xlabel("Receita de Bilheteria Mundial ($)")
plt.ylabel("Film")
plt.title("Top 10 Filmes da Pixar com Maior Bilheteria Mundial")
plt.show()
```

### <span style='background-color:#F97316; color:white; padding:5px; border-radius:5px;'>🏆 1. O Rei das Bilheteiras: Divertida Mente 2</span>
> O filme da Pixar com maior bilheteria, **Divertida Mente 2**, quebrou recordes com impressionantes **$1,69 bilhões arrecadados mundialmente!** 💰🎬

 *Um testemunho do poder narrativo e da profundidade emocional da Pixar, conquistando o público ao redor do mundo.*

```python
# Gráfico 2: ROI vs. Orçamento para todos os filmes da Pixar
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='budget', y='ROI', hue='film', palette='coolwarm', s=100)
plt.xlabel("Orçamento ($)")
plt.ylabel("ROI (Retorno sobre o Investimento)")
plt.title("ROI vs. Orçamento para os Filmes da Pixar")
plt.xscale('log')
plt.show()
```

### <span style='background-color:#10B981; color:white; padding:5px; border-radius:5px;'>💰 2. O Segredo para Lucros Máximos: Sucesso com Orçamento Médio</span>
> Surpreendentemente, **filmes com orçamentos moderados geram o maior ROI** — mostrando que gastar muito nem sempre garante grandes retornos.

 *Produções eficientes como* Toy Story *alcançaram enorme lucratividade com orçamento inteligente e forte apelo ao público.*

```python
# Gráfico 3: Tendências de Orçamento e Arrecadação nas Bilheteiras ao Longo do Tempo
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='release_date', y='budget', label='Orçamento', marker='o')
sns.lineplot(data=df, x='release_date', y='box_office_worldwide', label='Bilheteria Mundial', marker='o')
plt.xlabel("Ano de Lançamento")
plt.ylabel("Valor ($)")
plt.title("Orçamentos e Bilheterias dos Filmes da Pixar ao Longo do Tempo")
plt.legend()
plt.show()
```

### <span style='background-color:#3B82F6; color:white; padding:5px; border-radius:5px;'>📈 3. A Evolução da Estratégia Financeira da Pixar</span>
> Ao longo dos anos, **os orçamentos da Pixar aumentaram significativamente**, mas também **a arrecadação nas bilheteiras**, demonstrando uma notável **trajetória de crescimento**.

 *Essa tendência destaca o domínio da Pixar em equilibrar custos de produção com sucesso comercial, garantindo sustentabilidade no longo prazo.*

 *De* Toy Story (1995) *até* Divertida Mente 2 (2024), *a escala financeira da Pixar se transformou radicalmente.*

```python
# Gráfico 4.1: Distribuição das Avaliações da Crítica (Rotten Tomatoes e Metacritic)

df['imdb_score'] = df['imdb_score'] * 10 # A escala do IMDB é de 0 a 10

plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['rotten_tomatoes_score', 'metacritic_score', 'imdb_score']], palette='pastel')
plt.xlabel("Plataforma de Avaliação")
plt.ylabel("Nota")
plt.title("Distribuição das Avaliações dos Filmes da Pixar nas Plataformas")
plt.show()
```

### <span style='background-color:#9333EA; color:white; padding:5px; border-radius:5px;'>🍿 4. Um Legado de Reconhecimento da Crítica</span>
> A Pixar não é apenas sobre dinheiro — é sobre **excelência artística**. O estúdio frequentemente alcança **notas acima de 90% no Rotten Tomatoes**, comprovando seu talento narrativo incomparável.

 *Filmes da Pixar são mais do que sucessos de bilheteria — são fenômenos culturais que deixam marcas duradouras.*
