```python
# 1️⃣ Definir o caminho do vídeo no Drive (já montado)
video_path = "/content/drive/MyDrive/VID_3.mp4"  # ajuste o caminho se necessário

# 2️⃣ Carregar o modelo treinado (best.pt)
from ultralytics import YOLO

model = YOLO("/content/runs/detect/train/weights/best.pt")  # caminho do modelo treinado

# 3️⃣ Rodar a predição no vídeo
model.predict(video_path, save=True, conf=0.05)

# 4️⃣ Encontrar a pasta predict mais recente e o vídeo predito
import os
from google.colab import files

predict_base = "/content/runs/detect/"
predict_folders = [os.path.join(predict_base, d) for d in os.listdir(predict_base) if d.startswith("predict")]
predict_folders = sorted(predict_folders, key=os.path.getmtime, reverse=True)
latest_predict_folder = predict_folders[0]

files_in_folder = os.listdir(latest_predict_folder)

for file in files_in_folder:
    if file.endswith(".avi") or file.endswith(".mp4"):
        prediction_video = os.path.join(latest_predict_folder, file)
        break

# 5️⃣ Baixar o vídeo predito para o computador
files.download(prediction_video)
```
